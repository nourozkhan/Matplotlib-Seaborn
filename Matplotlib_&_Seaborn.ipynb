# **Matplotlib & Seaborn**
Matplotlib and Seaborn both are python library for data visuallization.

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

df = pd.read_excel("sample_data/linechart.xlsx")

df
```

```python
plt.figure(figsize=(10,5))
plt.plot(df.Quarter,df.Dishwasher, color='purple', label='Dishwasher')
plt.plot(df.Quarter,df["Washing Machine"], color='red', label='Washing Machine')
plt.plot(df.Quarter,df.Fridge, color='orange', label='Fridge')
plt.title("Production Report")
plt.xlabel("Time Line")
plt.ylabel("Production")
plt.legend()
plt.show()
```

```python
total_sales = df[['Fridge','Dishwasher','Washing Machine']].sum()
plt.pie(total_sales, labels=total_sales.index, autopct='%1.1f%%', shadow=True, explode=(0.1,0,0), startangle=140)
plt.title('Total Sales')
plt.show()
```

```python
histogram_df = pd.read_excel("sample_data/histograms.xlsx")
histogram_df.head()
```

```python
sns.histplot(histogram_df['Exam_Score'], kde=True)
plt.title('Exam Score')
plt.show()
```
